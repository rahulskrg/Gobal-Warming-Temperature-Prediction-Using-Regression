#### Variance Inflation Factor to check Multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif["features"] = X_train.columns
```

```python
vif.round(1)
```

Here CO2 and CH4 have a high variance inflation factor because they "explain" the same variance within this dataset. We would need to discard one of these variables before moving on to model building or risk building a model with high multicolinearity.

#### Model Evaluation

```python
from sklearn import metrics
```

```python
print('R^2:',metrics.r2_score(y_train, y_pred))
print('MAE:',metrics.mean_absolute_error(y_train, y_pred))
print('RSE:',metrics.mean_squared_error(y_train, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))

mape = np.mean(np.abs((y_train - y_pred) / y_train)) * 100
print('MAPE = ',mape)
```

69.37% of variation of dependent variable is expalined by independent variables.

Actual dependent variable deviates from regression line by 0.0095 units on average

#### R^2 calculation

```python
np.corrcoef(y_train,y_pred)
```

```python
R2=0.82809771*0.82809771
R2
```

```python

mean = np.mean(y_train)
sse = ((y_train-y_pred)**2).sum()
ssr = ((y_pred - mean)**2).sum()
sst = ((y_train-mean)**2).sum() ## sst,tss,tse
r2 = ssr/sst

n = len(y_train)
ar2 = 1 - ((1 -r2)*(n-1)/(n-4-1))

##OR, r2 = 1-(sse/sst)
print('SSE = ',sse)
print('SSR = ',ssr)
print('SST = ',sst)
print('R^2 = ',r2)
print('Adjusted R2=',ar2)
```
